````python
#@title utils
import os
import shutil
import uuid
from pydub import AudioSegment


def save_video(Image_Path,Audio_Path,video_folder,Face_Upscale):
  global base_path
  backup_folder=f"{base_path}/backup_video"
  if not os.path.exists(backup_folder):
    os.mkdir(backup_folder)
  test_image_name = os.path.splitext(os.path.basename(Image_Path))[0]
  audio_name = os.path.splitext(os.path.basename(Audio_Path))[0]
  predicted_video_256_path = f'{video_folder}/{test_image_name}-{audio_name}.mp4'
  predicted_video_512_path = f'{video_folder}/{test_image_name}-{audio_name}_SR.mp4'
  low=None
  high=None
  uuid_string = str(uuid.uuid4())[:6]

  if os.path.exists(predicted_video_256_path):
    shutil.move(predicted_video_256_path,f'{backup_folder}/{test_image_name}-{audio_name}_{uuid_string}.mp4')
    low=f'{backup_folder}/{test_image_name}-{audio_name}_{uuid_string}.mp4'
  if os.path.exists(predicted_video_512_path):
    if Face_Upscale:
      shutil.move(predicted_video_512_path,f'{backup_folder}/{test_image_name}-{audio_name}_SR_{uuid_string}.mp4')
      high=f'{backup_folder}/{test_image_name}-{audio_name}_SR_{uuid_string}.mp4'
  return low,high





def split_audio(input_file, output_folder, chunk_length_s=60):
    chunk_length_ms=chunk_length_s*1000
    # Ensure the output folder exists
    if os.path.exists(output_folder):
      shutil.rmtree(output_folder)
    os.makedirs(output_folder)

    # Load the audio file
    audio = AudioSegment.from_file(input_file)

    # Get the length of the audio in milliseconds
    audio_length_ms = len(audio)

    # Initialize the start time for the chunks
    start_time_ms = 0
    chunk_number = 1

    # Process each chunk
    while start_time_ms < audio_length_ms:
        # Calculate the end time for the chunk
        end_time_ms = min(start_time_ms + chunk_length_ms, audio_length_ms)

        # Extract the chunk
        chunk = audio[start_time_ms:end_time_ms]

        # Define the output file path
        output_file = os.path.join(output_folder, f'{chunk_number}.wav')

        # Export the chunk as a WAV file
        chunk.export(output_file, format='wav')

        # Move to the next chunk
        start_time_ms = end_time_ms
        chunk_number += 1


from pydub import AudioSegment
import os
import shutil

def split_audio_16k(input_file, output_folder, chunk_length_s=60, sample_rate=16000):
    chunk_length_ms=chunk_length_s*1000
    # Ensure the output folder exists
    if os.path.exists(output_folder):
        shutil.rmtree(output_folder)
    os.makedirs(output_folder)

    # Load the audio file
    audio = AudioSegment.from_file(input_file)

    # Convert the audio to the desired sample rate
    audio = audio.set_frame_rate(sample_rate)

    # Get the length of the audio in milliseconds
    audio_length_ms = len(audio)

    # Initialize the start time for the chunks
    start_time_ms = 0
    chunk_number = 1

    # Process each chunk
    while start_time_ms < audio_length_ms:
        # Calculate the end time for the chunk
        end_time_ms = min(start_time_ms + chunk_length_ms, audio_length_ms)

        # Extract the chunk
        chunk = audio[start_time_ms:end_time_ms]

        # Define the output file path
        output_file = os.path.join(output_folder, f'{chunk_number}.wav')

        # Export the chunk as a WAV file with the specified sample rate
        chunk.export(output_file, format='wav', parameters=["-ar", str(sample_rate)])

        # Move to the next chunk
        start_time_ms = end_time_ms
        chunk_number += 1


def make_npy_files(Audio_Path,npy_save_folder,Audio_Chunks_Duration_in_Seconds):
  global base_path
  split_audio_16k(Audio_Path, f'{base_path}/audios_16k',chunk_length_s=Audio_Chunks_Duration_in_Seconds)
  if os.path.exists(npy_save_folder):
    shutil.rmtree(npy_save_folder)
  os.mkdir(npy_save_folder)
  os.chdir(f"{base_path}/AniTalker")
  command=f'python {base_path}/AniTalker/extract_audio_features.py --model_path "{base_path}/AniTalker/ckpts/chinese-hubert-large" --audio_dir_path "{base_path}/audios_16k" --audio_feature_saved_path "{npy_save_folder}" --computed_device "cuda" --padding_to_align_audio True'
  print(f"!{command}")
  var=os.system(command)
  if var==0:
    print(f".npy file save at {npy_save_folder}")
  else:
    print(f"Failed to generate .npy files")



def concatenate_videos(Image_Path,Method):
  global base_path
  video_files=os.listdir(f"{base_path}/video_clips")
  sorted_video_filenames = sorted(video_files, key=lambda filename: int(filename.split('.')[0]))
  list_file=f"{base_path}/list.txt"
  with open(list_file, 'w') as f:
    for i in sorted_video_filenames:
      f.write(f"file '{base_path}/video_clips/{i}'\n")


  image_file_name = os.path.splitext(os.path.basename(Image_Path))[0]
  random_uuid = str(uuid.uuid4())[:6]
  method=Method.replace(" ","_")
  if not os.path.exists(f"{base_path}/save_video"):
    os.mkdir(f"{base_path}/save_video")
  output_path = f"{base_path}/save_video/{image_file_name}_{method}_{random_uuid}.mp4"

  ffmpeg_cmd=f"ffmpeg -f concat -safe 0 -i {list_file} -c copy {output_path} -y"
  var1=os.system(ffmpeg_cmd)
  if var1==0:
    print(f"Video save at {output_path}")
    if os.path.exists("/content/gdrive/MyDrive/"):
      f_name=os.path.basename(output_path)
      drive_save_path=f"/content/gdrive/MyDrive/video/{f_name}"
      shutil.copy(output_path,drive_save_path)
  else:
    print("Error")
    print(ffmpeg_cmd)
  return output_path

import hashlib
def compute_hash(filename):
    hasher = hashlib.sha256()
    with open(filename, 'rb') as f:
        while chunk := f.read(8192):
            hasher.update(chunk)
    return hasher.hexdigest()
import subprocess
def remove_silence_from_audio(audio_file, margin=0.1):
    base_path = os.path.dirname(audio_file)
    random_uuid = str(uuid.uuid4())[:6]
    new_save_path = None
    logs=None
    try:
        ffprobe_command = f"ffprobe -i {audio_file} -show_entries format=duration -v quiet -of csv=p=0"
        duration_string = subprocess.check_output(ffprobe_command, shell=True, text=True)
        duration = float(duration_string)
        # print(f"Audio duration {duration} seconds")

        ffmpeg_command = f"ffmpeg -f lavfi -t {duration} -i color=c=black:s=160x90 -c:v libx264 -strict experimental {base_path}/blank.mp4 -y"

        white_var = os.system(ffmpeg_command)

        if white_var == 0:
            response_var = os.system(f"ffmpeg -i {base_path}/blank.mp4 -i {audio_file} -map 0 -map 1 -c:v copy -c:a aac -strict experimental {base_path}/temp.mp4 -y")

            if response_var == 0:
                var = os.system(f"auto-editor {base_path}/temp.mp4 --margin {float(margin)}sec")

                if var == 0:
                    if not os.path.exists("/content/backup_video"):
                      os.mkdir("/content/backup_video")
                    new_save_path = f"/content/backup_video/{random_uuid}.wav"
                    final_var = os.system(f"ffmpeg -i {base_path}/temp_ALTERED.mp4 -vn -acodec pcm_s16le -ar 44100 -ac 2 {new_save_path} -y")
                    if final_var == 0:
                        ffprobe_command = f"ffprobe -i {new_save_path} -show_entries format=duration -v quiet -of csv=p=0"
                        after_duration_string = subprocess.check_output(ffprobe_command, shell=True, text=True)
                        after_duration = float(after_duration_string)
                        logs=f"""
                        Before: {duration} sec
                        \n
                        After: {after_duration} sec
                        """
                        print(f"Audio saved at {new_save_path}")
                        print(logs)

                    else:
                        print("Failed to remove silence from audio")
                else:
                    print("Failed to edit video")
            else:
                print("Can't add audio to blank video")
        else:
            print("Can't create blank video")

    except Exception as e:
        print(e)
        print("Can't find audio duration")
    return new_save_path,logs


#again stupid code for colab
from google.colab import files
import re
import os
import uuid
from IPython.display import clear_output
from google.colab import files

def clean_file_name(file_path):
  # Get the base file name and extension
  file_name = os.path.basename(file_path)
  file_name, file_extension = os.path.splitext(file_name)
  cleaned = re.sub(r'[^a-zA-Z\d]+', '_', file_name)   # Replace non-alphabetic characters with single underscore
  clean_file_name=re.sub(r'([a-zA-Z])_+([a-zA-Z])', r'\1_\2', cleaned)  # Replace multiple underscores between alphabets with one underscore
  clean_file_name = clean_file_name.strip('_')
  random_uuid = uuid.uuid4().hex[:6]
  # Combine cleaned file name with the original extension
  clean_file_path = os.path.join(os.path.dirname(file_path), clean_file_name +f"_{random_uuid}"+file_extension)
  return clean_file_path


def upload_image():
  upload_folder="/content/user_upload"
  if not os.path.exists(upload_folder):
    os.mkdir(upload_folder)
  os.chdir(upload_folder)
  uploaded = files.upload()
  os.chdir("/content/")
  image_name_list=[]
  for fn in uploaded.keys():
    file_path=f"{upload_folder}/{fn}"
    save_file_path=clean_file_name(file_path)
    os.rename(file_path,save_file_path)
    image_name_list.append(save_file_path)
  image_extensions = ('.png', '.jpg', '.jpeg')
  clear_output()
  for i in image_extensions:
    if image_name_list[-1].lower().endswith(i):
      return image_name_list[-1]
  else:
    print("Upload an Image")
  return None


def upload_audio():
  upload_folder="/content/user_upload"
  if not os.path.exists(upload_folder):
    os.mkdir(upload_folder)
  os.chdir(upload_folder)
  uploaded = files.upload()
  os.chdir("/content")
  clear_output()
  audio_name_list=[]
  for fn in uploaded.keys():
    file_path=f"{upload_folder}/{fn}"
    save_file_path=clean_file_name(file_path)
    os.rename(file_path,save_file_path)
    audio_name_list.append(save_file_path)
  audio_extensions = ('.wav', '.mp3')
  clear_output()
  for i in audio_extensions:
    if audio_name_list[-1].lower().endswith(i):
      return audio_name_list[-1]
  else:
    print("Upload an audio")
  return None

from IPython.display import HTML
from base64 import b64encode

def show_video(video_path, video_width = 600):

  video_file = open(video_path, "r+b").read()

  video_url = f"data:video/mp4;base64,{b64encode(video_file).decode()}"
  return HTML(f"""<video width={video_width} controls><source src="{video_url}"></video>""")

# end stupid code for colab



def AniTalker_colab_interface(Method,Image_Path,Raw_Audio_Path,Chunks_Duration,Face_Upscale,Remove_Slience=False):
  global base_path,old_audio_path
  video_folder=f"{base_path}/download_video"
  video_clips_folder=f"{base_path}/video_clips"
  npy_save_folder=f"{base_path}/AniTalker/hubert_features"
  Hubert_Support=["Hubert Full Control","Hubert Pose Only","Hubert Audio Only","MFCC Pose Only",]
  if Remove_Slience:
    Audio_Path,_=remove_silence_from_audio(Raw_Audio_Path)
  else:
    Audio_Path=Raw_Audio_Path
  time_save = False
  if old_audio_path:
      # print(old_audio_path,Audio_Path)
      hash1 = compute_hash(old_audio_path)
      hash2 = compute_hash(Audio_Path)
      # print(f"Hash1: {hash1}")
      # print(f"Hash2: {hash2}")
      if hash1 == hash2:
          time_save = True
          print("The audio files are same")
      else:
        # print("The audio files are differnet")
        pass

  if Method != "MFCC Full Control":
      old_audio_path = Audio_Path

  if Method in Hubert_Support and not time_save:
      print("Calling make_npy_files")
      make_npy_files(Audio_Path, npy_save_folder, Chunks_Duration)
  else:
      print(f"Skipping make_npy_files - time_save: {time_save}")

  if os.path.exists(video_clips_folder):
    shutil.rmtree(video_clips_folder)
  os.mkdir(video_clips_folder)
  split_audio(Audio_Path, f'{base_path}/audio_clips',chunk_length_s=Chunks_Duration)
  audio_files=os.listdir(f"{base_path}/audio_clips")
  sorted_filenames = sorted(audio_files, key=lambda filename: int(filename.split('.')[0]))
  Seed=len(sorted_filenames)+100000


  for i in sorted_filenames:
    temp_audio_path=f"{base_path}/audio_clips/{i}"
    if Method in Hubert_Support:
      np_file=i.replace("wav","npy")
      hubert_file=f"{npy_save_folder}/{np_file}"
    else:
      hubert_file=""
    command=get_command(Method,Image_Path,temp_audio_path,video_folder,hubert_file,Seed,Face_Upscale)
    print(command)
    var=os.system(command)
    if var==0:
      low_resolution_path,high_resolution_path=save_video(Image_Path,temp_audio_path,video_folder,Face_Upscale)
      v_name=i.replace("wav","mp4")
      if high_resolution_path!=None and Face_Upscale==True:
        shutil.copy(high_resolution_path,f"{video_clips_folder}/{v_name}")
      elif low_resolution_path!=None:
        shutil.copy(low_resolution_path,f"{video_clips_folder}/{v_name}")
      else:
        pass
    else:
      print(f"Failed")
      print(f"!{command}")

  output_path=concatenate_videos(Image_Path,Method)
  return output_path
````
